```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
body_df = pd.read_csv('Downloads/body.csv',index_col=0)
body_df.head()
```

```python
body_df['Height'] # Pandas Series
```

```python
body_df['Height'].value_counts().sort_index() # 빈도 순으로 정렬 / index 순으로 정렬
```

```python
body_df['Height'].value_counts().sort_index().plot()
```

```python
sns.kdeplot(body_df['Height']) # 위에꺼가 아래처럼 부드러운 곡선화 됨
```

```python
sns.kdeplot(body_df['Height'], bw=0.5)
# bw는.. 히스토그램으로 치면 몇개의 구간으로 나눌지와 같은 뜻
# bw가 너무 커지면 상당히 부정확해지므로 적당히 쓸 줄 알아야됨
```

```python
body_df.head()
```

```python
body_df.plot(kind='hist', y='Height', bins=15)
```

```python
sns.distplot(body_df['Height'], bins=15)
#distplot 이용 시 히스토그램 위에 kde가 덮어씌워짐
```

```python
body_df.plot(kind='box', y='Height')
```

```python
sns.violinplot(y=body_df['Height'])
#kde 그래프를 대칭되게 그린 것. 특정 분포에는 바이올린처럼 생겼다고 해서 violinplot임
```

```python
body_df.plot(kind='scatter', x='Height', y='Weight')
#175cm에 70kg정도 되는 애들이 많다. (점이 매우많음))
```

```python
sns.kdeplot(body_df['Height'], body_df['Weight'])
```

```python
sns.kdeplot(body_df['Height'])
```

```python
sns.kdeplot(body_df['Weight'])
```

```python
# LM Plot
sns.lmplot(data=body_df, x='Height', y='Weight')
# 데이터 파라미터에 데이터프레임
# X, Y축에 넣고 싶은 column 이름
# 선 -> regression line(회귀선)
```

```python
# 카테고리별 시각화
laptops_df = pd.read_csv('Downloads/laptops.csv')
laptops_df.head()
```

```python
laptops_df['os'].unique()
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='box')
# 박스플롯의 단점은.. 각 데이터가 몇개씩 있는지 모른다. 이럴 때는 strip plot을 쓰면 됨
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='strip')
```

```python
laptops_df['processor_brand'].unique()
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='strip', hue='processor_brand')
# 점이 겹친다는 문제가 생김
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='swarm', hue='processor_brand')
# swarm plot을 써주면 점들이 겹치지 않음
```
