```python
import pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Desktop/DS-titanic/train.csv')
df
```

```python
# 1. Pclass
# Pclass에 대해서 적용해볼 수 있는 건 성별, 나이, 요금, 선착장? 정도가 있을 것 같다.
```

```python
df[df['Pclass'] == 3]
# 전체 인원은 891명이고, 그 중 3등석 인원은 491명이다.
```

```python
df[(df['Pclass'] == 3) & (df['Sex'] == 'male')]
# 3등석 인원 491명 중 남자인 사람은 347명, 약 70%정도이다. 그렇다면 여자의 비율은 당연히 30% 정도가 될 것.
```

```python
df[df['Pclass'] == 2]
```

```python
df[(df['Pclass'] == 2) & (df['Sex'] == 'male')]
# 2등석 184명 중 남자 108명. 역시 60%정도의 비율이다.
```

```python
df[df['Pclass'] == 1]
```

```python
df[(df['Pclass'] == 1) & (df['Sex'] == 'male')]
# 1등석 216명 중 남자 비율 122명. 역시 55% 이상이다. 이로서, 3,2,1등석 모두 남자의 비율이 좀 더 높은 것을 알 수 있다.
```

```python
sns.countplot(data=df, x="Pclass", hue="Survived")
# 3등석에서 1등석으로 갈 수록 사망의 비율이 낮아지고, 3->2에서 좀 줄어들긴 하지만, 3에서 1로 가면서 생존자의 비율도 늘어난다.
```

```python
sns.countplot(data=df, x='Pclass', hue='Embarked')
# S의 비율이 제일 높고, Q 선착장의 경우 3등석이 대부분이다.
```

```python
# 2. Sex
# 성별과 비교할 수 있는 건.. 일단 생존률?
sns.countplot(data=df, x="Sex", hue="Survived")
# 여성의 생존률이 남성보다 대충 봐도 훨씬 높다는 것을 알 수 있다.
# 생존률 외에 다른 게 딱히 생각나는 게 없다..
```

```python
# 3. Age
# 나이와 Pclass, 성별 등을 비교할 수 있을 것 같다.
sns.catplot(data=df, x='Sex', y='Age', kind='strip')
# 일단 남성의 비율이 더 많은 것을 알 수 있고, 모수가 많은 만큼 연령대 분포도 남자 쪽이 더 넓게 퍼져있음을 알 수 있다.
```

```python
sns.catplot(data=df, x='Pclass', y='Age', kind='swarm', hue='Sex')
# 일단 당연히 1,2등석에 비해 3등석에 사람이 많고, 1등석에 여자의 비율이 더 높음을 알 수 있다. 그리고, 고연령층은 남녀 가릴 것 없이 1등석에 있는 비율이 높았다.
```

```python
sns.catplot(data=df, x='Pclass', y='Fare', kind='swarm', hue='Sex')
# 1등석으로 갈 수록 탑승자들이 낸 돈의 평균이 점점 높아지긴 했다. 근데 3등석과 2등석은 큰 차이는 보이지 않았다.
# 그리고, 1등석에서도 2,3등석에서와 비슷한 요금을 낸 사람도 있었고, 이런 데이터보다 중요한 건 전체적으로 봤을 때 역시 1등석으로 갈 수록
# 비용이 높았다는 것이다.
```

```python
# Embarked(선착장) -> 요금, 성별 등등 다양하게 비교해볼 수 있을 것 같다.
sns.catplot(data=df, x='Embarked', y='Fare', kind='swarm', hue='Sex')
# 1. C 선착장에서 일부가 매우 눞은 요금을 냈음 (사실 이건 그렇게 중요한 건 아닌 것 같다)
# 2. S > C > Q 순으로 많음. Q 선착장에서 탑승한 사람들의 평균 요금이 제일 낮은 것을 보아 3등석이 제일 많을 것이라는 추론을 할 수 있다.
# S의 평균 비용이 제일 높다. 이것을 통해 대충 S와 C는 비슷한 경제 수준 > Q는 S와 C보다 못한 경제수준에 있음을 짐작할 수 있다.
```
