```python
import pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Downloads/iris.csv')
df
```

```python
# 이 데이터는, 각 꽃 별 sepal_length(꽃받침 길이), sepal_width(꽃받침 폭), petal_length(꽃잎 길이), petal_width(꽃잎 폭)을
# 나타내었다. 이 데이터에 대한 추측을 해보자면 ..
# 1. 종 별로 꽃받침이나 꽃잎의 성질 중 뚜렷하게 구분되는 것이 있을 것이다. (종이 여러개고 각 종마다 생김새가 다 다르므로)
# 2. 꽃받침과 꽃잎의 길이, 폭은 각각 양의 상관관계를 가질 것이다. 정도가 있을 것 같다.
```

```python
# 1. 각 종 별 꽃받침이나 꽃잎 등의 차이
# 1-1. 각 종 별 꽃받침 길이의 차이
sns.catplot(data=df, x='species', y='sepal_length')
```

```python
# 해당 그래프가 시사하는 바는 매우 뚜렷하다. setosa의 평균 꽃받침 길이가 제일 짧고, virginica의 평균 꽃받침 길이가 제일 길다.
```

```python
# 1-2. 각 종 별 꽃받침 폭의 차이
sns.catplot(data=df, x='species', y='sepal_width')
```

```python
# 해당 그래프를 통해서는, setosa의 평균 꽃받침 폭이 나머지 두 종에 비해 확실히 길다는 것과,
# 꽃받침 폭 데이터만으로는 versicolor species와 virginica를 구분하기 어렵다는 것을 알 수 있다.
```

```python
# 1-3. 각 종 별 꽃잎 길이의 차이
sns.catplot(data=df, x='species', y='petal_length')
```

```python
# 해당 그래프는 꽃받침의 그것보다 훨씬 더 시사하는 바가 명확하다. setosa의 꽃잎 길이는 매우 짧은 편에 속하고, virginica의 꽃잎 길이가 제일 길다.
# 이상점이 거의 존재하지 않는 것으로 보아 신빙성이 높다는 것을 추론할 수 있다.
```

```python
# 1-4. 각 종 별 꽃잎 폭의 차이
sns.catplot(data=df, x='species', y='petal_width')
```

```python
# 해당 그래프 또한 꽃받침의 그것보다 시사하는 바가 명확하다. setosa의 꽃받침 길이가 제일 짧고 virginica의 꽃받침 길이가 제일 길다.
```

```python
# 이 네 가지의 그래프를 통해 알 수 있었던 사실은 ..
# 1. '꽃받침 폭' 만으로 setosa인지의 여부를 알 수 있다. 반대로 말하면, '꽃받침 폭'으로는 나머지 두 개를 구별하긴 어렵다.
# 2. (종의 모수가 부족하다는 문제점이 있으나 3개로 한정하자면) 꽃잎 길이와 꽃잎 폭은 양의 상관관계를 갖고 있다.
# 3. 꽃받침의 길이와 폭은 특정한 상관관계를 띠고 있다고 보기 어렵다.
```
